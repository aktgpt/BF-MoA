```python
import pandas as pd
import numpy as np
import random
import cv2
from functions import *
from math import ceil
```

```python
csv_path = '/scratch-shared/phil/SPECS_phil/exp_stats/bf_main.csv'
df = pd.read_csv(csv_path)
```

```python
# ten selected MOAs
moas = ['Aurora kinase inhibitor', 'tubulin polymerization inhibitor', 'JAK inhibitor', 'protein synthesis inhibitor', 'HDAC inhibitor', 
        'topoisomerase inhibitor', 'PARP inhibitor', 'ATPase inhibitor', 'retinoid receptor agonist', 'HSP inhibitor']
```

```python
df_sub = df[df['moa'].isin(moas)].reset_index(drop=True)
df_sub['moa'].value_counts()
```

```python
df_sub
```

```python
df_sub.groupby(['moa', 'compound']).count()
```

```python
while True:
    df_sub = df_sub.sample(frac=1).reset_index(drop=True)
    train_df_unq_plate_list=[]
    
    n_splits = 5
    for split in range(n_splits):
        test_compounds = []
        for m in df_sub.moa.unique():
            comp_list = list(df_sub[df_sub.moa == m].compound.unique())

            chunked_list = list()
            chunks = [int(np.floor(len(comp_list) / n_splits))] * 4
            chunks.append(len(comp_list) - sum(chunks))

            prev_chunk = 0
            for i in range(len(chunks)):
                chunked_list.append(comp_list[prev_chunk:prev_chunk + chunks[i]])
                prev_chunk += chunks[i]

            test_compounds.append(chunked_list[split])

        test_compounds = sum(test_compounds, [])
        test_df = df_sub[df_sub.compound.isin(test_compounds)].reset_index(drop=True)
        train_full_df = df_sub[~df_sub.compound.isin(test_compounds)].reset_index(drop=True)

        val_compounds = []
        train_compounds = []
        for m in train_full_df.moa.unique():
            comp_list = list(train_full_df[train_full_df.moa == m].compound.unique())
            val = random.sample(comp_list, round(len(comp_list) * 0.2))
            train = list(set(comp_list) - set(val))
            val_compounds.append(val)
            train_compounds.append(train)

        val_compounds = sum(val_compounds, [])
        train_compounds = sum(train_compounds, [])

        print('split' + str(split + 1) + ': comps train = ' + str(len(train_compounds)) +
              '; val = ' + str(len(val_compounds)) + '; test = ' + str(len(test_compounds)))

        val_df = df_sub[df_sub.compound.isin(val_compounds)].reset_index(drop=True)
        train_df = df_sub[df_sub.compound.isin(train_compounds)].reset_index(drop=True)
        train_df_unq_plate_list.append(len(train_df.plate.unique()))
        
        train_df.to_csv('stats/new_train_val_test_5_splits/bf_subset_train_split' + str(split + 1) + '.csv', index=False)
        val_df.to_csv('stats/new_train_val_test_5_splits/bf_subset_val_split' + str(split + 1) + '.csv', index=False)
        test_df.to_csv('stats/new_train_val_test_5_splits/bf_subset_test_split' + str(split + 1) + '.csv', index=False)
    
    print(train_df_unq_plate_list)
    plate_18_count = len([x for x in train_df_unq_plate_list if x==18])
    
    if plate_18_count ==3:
        break
    
```
