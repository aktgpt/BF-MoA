```python
import random
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import cv2
import re
import glob
import os, shutil
import datetime
import xml.etree.ElementTree as ET

# misc functions
def atoi(text):
    return int(text) if text.isdigit() else text

def natural_keys(text):
    return [atoi(c) for c in re.split('(\d+)', text)]
```

```python
plate_ids = []
for i in np.arange(15076, 15100):
    plate_ids.append('P0' + str(i))
```

```python
plate_barcodes = pd.DataFrame({'layout_id': sorted(list(set(plates_all['layout_id']))),
                              'barcode': plate_ids})
```

```python
plate_barcodes = plate_barcodes.loc[plate_barcodes['layout_id'].isin(layout_ids)].reset_index(drop=True)
```

```python
plate_barcodes
```

```python
root_dir = '/mnt/micro-images-pvc/specs935-v1/'
```

```python
def what_channel(file, c1, c2, c3, c4, c5):
    file.split('_')[-1].split('-')[0]
    if 'w1' in file:
        c1.append(file)
        return c1,c2,c3,c4,c5
    elif 'w2' in file:
        c2.append(file)
        return c1,c2,c3,c4,c5
    elif 'w3' in file:
        c3.append(file)
        return c1,c2,c3,c4,c5
    elif 'w4' in file:
        c4.append(file)
        return c1,c2,c3,c4,c5
    elif 'w5' in file:
        c5.append(file)
        return c1,c2,c3,c4,c5
    else:
        return None
```

```python
plate_list = []
well_list = []
moa_list = []
compound_list = []
path_list = []
c1 = []
c2 = []
c3 = []
c4 = []
c5 = []
for barcode, layout in zip(plate_barcodes['barcode'], plate_barcodes['layout_id']):
    image_folder = sorted(glob.glob(root_dir + '*' + barcode + '*/*/*'))[-1]
    wells = list(plates[plates['layout_id']==layout]['well_id'])
    f0 = sorted(glob.glob(image_folder + '/*'), key=natural_keys)
    for well in wells:
        f1 = sorted([x for x in f0 if 'thumb' not in x and '_' + well in x])
        batch_id = list(plates.loc[(plates['layout_id']==layout) & (plates['well_id']==well)]['batch_id'])[0]
        compound = list(compounds.loc[(compounds['Batch nr']==batch_id)]['Compound ID'])[0]
        moa = list(compounds.loc[(compounds['Batch nr']==batch_id)]['moa'])[0]
        
        for file in f1:
            c1,c2,c3,c4,c5 = what_channel(file.split('/')[-1], c1, c2, c3, c4, c5)
        
        plate_list.append([barcode for i in range(9)])
        well_list.append([well for i in range(9)])
        moa_list.append([moa for i in range(9)])
        compound_list.append([compound for i in range(9)])
        path_list.append([image_folder for i in range(9)])
        #moas.append(moa)
        #compound_list.append(compound)
        
plate_list = sum(plate_list, [])
well_list = sum(well_list, [])
moa_list = sum(moa_list, [])
compound_list = sum(compound_list, [])
path_list = sum(path_list, [])
    
```

```python
df = pd.DataFrame(
    {
        'plate': plate_list,
        'well': well_list,
        'compound': compound_list,
        'path': path_list,
        'C1': c1,
        'C2': c2,
        'C3': c3,
        'C4': c4,
        'C5': c5,
        'moa': moa_list
    }
)
```

```python
df.to_csv('~/SPECS_phil/exp_stats/main.csv', index=False)
```
