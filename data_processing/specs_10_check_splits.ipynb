```python
import numpy as np
import pandas as pd
```

```python
for split in range(5):
    for sub in ['train', 'val', 'test']:
        fl = pd.read_csv('stats/new_train_val_test_5_splits/f_subset_' + sub + '_split' + str(split + 1) + '.csv')
        bf = pd.read_csv('stats/new_train_val_test_5_splits/bf_subset_' + sub + '_split' + str(split + 1) + '.csv')

        fl_pwcm = fl.plate + fl.well + fl.compound + fl.moa
        fl_pwcm = np.sort(np.unique(fl_pwcm))
        bf_pwcm = bf.plate + bf.well + bf.compound + bf.moa
        bf_pwcm = np.sort(np.unique(bf_pwcm))

        if not np.array_equal(fl_pwcm, bf_pwcm): 
            print('False!')
```
