```python
import pandas as pd
import numpy as np
```

```python
for i in range(1, 6):
    bf_path = '/home/jovyan/SPECS_phil/stats/new_train_val_test_5_splits/bf_subset_train_split' + str(i) + '.csv'
    #bf_path = '/scratch-shared/phil/SPECS_phil/exp_stats/bf_main.csv'
    #f_path = '/scratch3-shared/phil/tiles_main.csv'
    f_path = '/scratch-shared/phil/SPECS_phil/exp_stats/main.csv'

    bf_df = pd.read_csv(bf_path)
    df = pd.read_csv(f_path)

    df2 = df[df.plate.isin(bf_df.plate.unique())]
    df2 = df2[df2.well.isin(bf_df.well.unique())]
    df2 = df2[df2.moa.isin(bf_df.moa.unique())]
    df2 = df2[df2.compound.isin(bf_df.compound.unique())]
    
    print('train split ' + str(i))
    print(df2.moa.value_counts())
    df2.to_csv('stats/new_train_val_test_5_splits/f_subset_train_split' + str(i) + '.csv', index=False)
```
